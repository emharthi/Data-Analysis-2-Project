```python
import pandas as pd

chunk_size = 10000  # عدد الأسطر التي سيتم قراءتها في كل دفعة
chunks = []

with pd.read_json('yelp_academic_dataset_review.json', lines=True, chunksize=chunk_size) as reader:
    for chunk in reader:
        chunks.append(chunk)
        # يمكنك معالجة كل دفعة هنا إذا كنت بحاجة إلى ذلك لتوفير الذاكرة
        # على سبيل المثال: تحليل البيانات أو تخزينها في ملف آخر

df = pd.concat(chunks, ignore_index=True)
print(df.head())
```
