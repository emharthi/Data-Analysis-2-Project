***Data Processing***

```python
#After that, we will define feature (X:all columns) and target (y:Target variable (Diagnosis)) variables,
# and split the dataset into training and testing sets.
from sklearn.model_selection import train_test_split

X = df.drop('Diagnosis', axis=1) # we drop the Diagnosis column because this is the Target
y = df['Diagnosis']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```

***Model Building and Training***

```python
# Bulid the model
from sklearn.naive_bayes import GaussianNB

model = GaussianNB() # we will use Gaussian Naive Bayes

model.fit(X_train, y_train); # training the model
```

***Evaluation the Model***

```python
# import some libraries
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)
```

***Cross-Validation***

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.naive_bayes import GaussianNB

X = df.drop('Diagnosis', axis=1) # we drop the Diagnosis column because this is the Target
y = df['Diagnosis']

k_folds = KFold(n_splits = 4)

scores = cross_val_score(model, X, y, cv = k_folds)

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```

```python
y_pred = model.predict(X_test)

accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

```python
labels = ["maligant", "benign"]
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot();
```

```python
print('\nClassification Report:')
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt


y_prob = model.predict_proba(X_test)[:, 1]


fpr, tpr, thresholds = roc_curve(y_test.map({'M': 1, 'B': 0}), y_prob)


plt.plot(fpr, tpr, label='Naive Bayes (AUC = {:.2f})'.format(roc_auc_score(y_test.map({'M': 1, 'B': 0}), y_prob)))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```
